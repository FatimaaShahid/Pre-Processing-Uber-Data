```python
import numpy as np
import pandas as pd
```

```python
PATH = '/content'
data=pd.read_csv(f'{PATH}/ncr_ride_bookings.csv')
#Displaying the read contents
data
```

```python
data1 = data.drop(["Reason for cancelling by Customer","Cancelled Rides by Driver","Driver Cancellation Reason","Incomplete Rides","Incomplete Rides Reason","Cancelled Rides by Customer"],axis=1)
```

```python
data1
```

```python
data2 = data1.dropna()

data2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data2["Booking Status"].unique()
```

```python
data2["Vehicle Type"].unique()
```

```python
data3 = pd.get_dummies(data2,columns=["Booking Status","Vehicle Type","Pickup Location","Drop Location","Payment Method"],dtype=int)
```

```python
data3
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data3['Avg VTAT'], bins=20, kde=True)
plt.title("AVG VTAT")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data3['Avg CTAT'], bins=30, kde=True)
plt.title("AVG CTAT")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data3['Booking Value'], bins=30, kde=True)
plt.title("Booking Value")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data3['Ride Distance'], bins=30, kde=True)
plt.title("Ride Distance")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data3['Driver Ratings'], bins=30, kde=True)
plt.title("Driver Ratings")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data3['Customer Rating'], bins=30, kde=True)
plt.title("Customer Rating")
plt.show()




```

```python
bins = [1,2,3,4,5]
labels = ["poor","fair","good","best"]

data3["binned_driver_rating"] = pd.cut(data3["Driver Ratings"],bins=bins, labels=labels)
data3
```

```python
from sklearn.preprocessing import MinMaxScaler
cols_to_normalize = ['Avg VTAT', 'Avg CTAT', 'Booking Value', 'Ride Distance']
scaler = MinMaxScaler()
data3[cols_to_normalize] = scaler.fit_transform(data3[cols_to_normalize])

data3
```
